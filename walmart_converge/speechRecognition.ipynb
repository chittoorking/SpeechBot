```python
connection_uri = "sqlite:///C:/SQLite_CustomerAnalysisData/stores.db" 
db_engine = sqlalchemy.create_engine(connection_uri)
```

```python
def extract_unique_productline_to_pandas(tablename, db_engine):
        query = '''SELECT DISTINCT(productLine) 
                   FROM {} ;'''.format(tablename)
        print(pd.read_sql(query, db_engine))

extract_unique_productline_to_pandas('products',db_engine)

Generate_text_to_speech("which product line you are looking for from the given below?")

productLine=RecordUserInput.recordAudio()
```

```python
def extract_productline_output_to_pandas(tablename1,tablename2,db_engine,productLine):
        query = f'''SELECT {tablename1}.productCode as productCode,productName,buyPrice,COUNT(quantityOrdered) as ordersReceived
                   FROM {tablename1} 
                   JOIN {tablename2}
                   ON {tablename1}.productCode=={tablename2}.productCode
                   WHERE UPPER(productLine)==UPPER("{productLine}")
                   GROUP BY {tablename1}.productCode;'''
        return pd.read_sql(query, db_engine)
productLine_df=extract_productline_output_to_pandas("products","orderDetails",db_engine,productLine)
productLine_df
```

```python
Generate_text_to_speech("What is the maximum price you want?")
time.sleep(1)
max_price= RecordUserInput.recordAudio()
```

```python
try:
    max_price=w2n.word_to_num(max_price)
except:
    pass 
productLine_df=productLine_df[productLine_df['buyPrice']<=max_price]
productLine_df
```

```python
responses=["I'm sorry ! I couldn't find anything like that",
 '{} is a great choice!',
 '{} or {} would work!',
 '{} is one option, but I know others too ']
df = recommendProducts(productLine_df,"ordersReceived")
df
```

```python
names = df.productName
n= min(len(df),3)
    # Select the nth element of the responses array
print(responses[n].format(*names))
Generate_text_to_speech(responses[n].format(*names))
time.sleep(1)
Generate_text_to_speech("Is it what you are looking for ?")
time.sleep(1)
reply=RecordUserInput.recordAudio()
```
